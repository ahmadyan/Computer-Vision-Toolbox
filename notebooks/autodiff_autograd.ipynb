# autodiff example using autograd
autograd imports a thinly-wrapped numpy, then we can define our own function. More examples are available [Here](https://github.com/HIPS/autograd/tree/master/examples)

```python
import autograd.numpy as np
from autograd import grad, value_and_grad 
```

```python
def tanh(x):
    y = np.exp(-2.0 * x)
    return (1.0 - y) / (1.0 + y)

def rozenbrock(x, y):
    return (1-x)**2 + 100*(y - x**2)**2
```

```python
from autograd import elementwise_grad as egrad
%matplotlib inline  
import matplotlib.pyplot as plt

x = np.linspace(-7, 7, 200)
plt.plot(x, tanh(x), x, egrad(tanh)(x), x, egrad(egrad(tanh))(x))
plt.show()
```

```python
def rosenbrock(x):
    return 100*(x[1] - x[0]**2)**2 + (1 - x[0])**2
```

```python
from scipy.optimize import minimize

rosenbrock_with_grad = value_and_grad(rosenbrock)

# Optimize using conjugate gradients.
result = minimize(rosenbrock_with_grad, x0=np.array([0.0, 0.0]), jac=True, method='CG')
print("Found minimum at {0}".format(result.x))
```
